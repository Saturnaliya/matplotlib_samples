```python
penguins = sns.load_dataset('penguins')
```

```python
cols = ["species", "island", "bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g", "sex"]
g = sns.PairGrid(penguins[cols])
g.map(sns.scatterplot)
g.map_diag(sns.histplot)
```

```python
penguins.groupby(["bill_length_mm", "body_mass_g"]).agg ({"flipper_length_mm": "max"}).plot(kind = "bar")
```

```python
pg.loc[pg['bill_length_mm']< 35, 'height_group'] = 'small'
pg.loc[(pg['bill_length_mm']>= 35) & (pg['bill_length_mm']<42),'height_group'] = 'middle'
pg.loc[pg['bill_length_mm']>42, 'height_group'] = 'big'
```

```python
pg
```

```python
sns.histplot(data = pg, x = 'flipper_length_mm', hue = 'height_group')
```

```python
def func(penguins):
  val=None
  if penguins > 42:
    val='high'
  elif 35 <= penguins >= 42:
    val = 'middle'
  elif penguins < 35:
    val = 'low'
  return val

  penguins['height_group']= penguins['bill_length_mm'].apply(func)

```
